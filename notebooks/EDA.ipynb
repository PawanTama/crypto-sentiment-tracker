Basic exploratory analysis for merged crypto price and sentiment data.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/merged_crypto_data.csv', parse_dates=['datetime'])
df.head()
```

```python
# Summary statistics

df.describe()
```

```python
# Check for nulls
df.isnull().sum()
```

```python
# Plot last 100 rows
sample_df = df.tail(100)
plt.figure(figsize=(10, 4))
plt.plot(sample_df['datetime'], sample_df['sentiment_score'], color='orange')
plt.title('Sentiment Score â€“ Last 100 Records')
plt.xlabel('Datetime')
plt.ylabel('Sentiment Score')
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
# Line plot: Bitcoin Price over Time
plt.figure(figsize=(12, 5))
plt.plot(df['datetime'], df['price'], color='green')
plt.title('Bitcoin Price Over Time')
plt.xlabel('Datetime')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

```python
# Scatterplot: Sentiment vs. % Price Change
# If % price change column exists
if 'Price Change' in df.columns:
    plt.figure(figsize=(8, 5))
    plt.scatter(df['sentiment_score'], df['Price Change'], alpha=0.6)
    plt.title('Sentiment Score vs. Price Change')
    plt.xlabel('Sentiment Score')
    plt.ylabel('% Price Change')
    plt.grid(True)
    plt.show()
else:
    print("Price Change column not found. Please calculate it in the processing script.")
```
